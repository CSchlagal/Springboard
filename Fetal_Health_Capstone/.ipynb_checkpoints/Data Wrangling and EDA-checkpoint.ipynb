Goal will be to create a model to classify the outcome of Cardiotocogram (CTG) exam which are found in the dataset and which represents the well being of the fetus. Dataset comes from Ayres de Campos et al. (2000) SisPorto 2.0 A Program for Automated Analysis of Cardiotocograms. J Matern Fetal Med 5:311-318

```python
import pandas as pd
import numpy as np
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fetal_health = "fetal_health.csv"
df = pd.read_csv(fetal_health)
df.head()
```

```python
df.describe()
```

```python
df.info()
```

There doesn't appear to be any missing values from the data set(need to double check) and they are all in numerical (float) values. 

```python
df.isnull().any()
```

```python
df.isna().any()
```

```python
df['fetal_health'] = df['fetal_health'].map({1: 'Normal', 2: 'Suspect', 3: 'Pathological'})
```

```python
plt.figure(figsize = (10,6))
ax = sns.countplot(x = df['fetal_health'])
plt.xlabel('Fetal Outcomes')
plt.ylabel('Counts')
plt.title('Total Counts by Fetal Outcome Score');
```

```python
df_f = df.select_dtypes(exclude='object')
plt.figure(figsize = (12,8))
sns.heatmap(data = df_f.corr());
```
